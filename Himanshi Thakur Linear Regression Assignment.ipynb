### STEP 1 - Reading and understanding the Data;

```python
# Importing necessary Libraries;

import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
```

```python
# Reading the CSV file;

df = pd.read_csv("day.csv")
df.head()
```

```python
# checking the size of dataframe;

df.shape
```

```python
# Calling for the standard description of the dataframe;

df.describe()
```

```python
# Fetching the basic information about the dataframe;

df.info()
```

```python
# Checking for the Null values in the dataframe;

df.isnull().sum()
```

```python
bike_df = df    # creating a new dataframe with similar values as "df" as to avoid any permanent changes in original dataframe.
```

```python
# Dropping the duplicates in dataframe, if any;

bike_df.drop_duplicates(inplace=True)
```

```python
bike_df.shape
```

- Apparently there weren't any duplicates as the shape of our dataframe remains unchanged as you can see in the cell above.

```python
bike_df.columns
```

- In the following cell we are renaming the column names for better understanding. For example:- "cnt" to "Count", etc.

```python
bike_df.rename(columns={"yr" : "Year", "mnth":"Month", "hum":"Humidity", "cnt":"Count"}, inplace=True)
```

```python
bike_df.head() # Checking if the changes took place;
```

```python
# Rounding the Decimal points upto 3 for better understanding and to make the data look more presentable;

col_list = ["temp", "atemp", "Humidity", "windspeed"]
bike_df[col_list] = round(bike_df[col_list], 3)
```

```python
bike_df.head()           # to check if the changes took place;
```

- In the following cell we will be changing the cetegorical values from numerical labels to descriptive representation.

```python
def func_season(A):
    return A.map({1 : "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter"})
bike_df["season"]=bike_df[["season"]].apply(func_season)
```

```python
def func_Month(A):
    return A.map({1:"January",2:"February",3:"March", 4:"April", 5:"May" , 6:"June", 7:"July", 8:"August", 9:"Sept" , 10:"October" , 11:"Nov" , 12:"December"})
bike_df["Month"]=bike_df[["Month"]].apply(func_Month)
```

```python
def func_Weather(A):
    return A.map({1:"Good Weather",2:"Cloudy",3:"Light Snow", 4:"Heavy Rain"})
bike_df["weathersit"] = bike_df[["weathersit"]].apply(func_Weather)
```

### STEP 3 - Data Preparation;

#### Concatenating all Dummy Dataframes with original Dataframe

```python
bike_df = pd.concat([bike_df, season_dataframe, month_dataframe, weather_dataframe], axis = 1)
bike_df.head()
```

```python
bike_df.columns    # Cheacking for all existing columns;
```

- Deleting/Dropping all non essential/duplicate columns.

```python
bike_df.drop(["season", "Month", "weathersit"], axis = 1, inplace=True)
```

### Step 4 - Splitting the data : Training and Testing;

### Step 5 - Building a Model;

-  Using LinearRegression Function from SciKit Learn as it is compatible with RFE.

```python
# Importing required Libraries;

from sklearn.feature_selection import RFE

from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
LR.fit(train_X, train_Y)

# Now we will be using/running RFE;

RFE = RFE(LR, 14)           
RFE = RFE.fit(train_X, train_Y)
```

```python
list(zip(train_X.columns, RFE.support_, RFE.ranking_))
```

```python
true_values = train_X.columns[RFE.support_]          # Figuring out TRUE RFE values;
true_values
```

```python
false_values = train_X.columns[~RFE.support_]        # Figuring out FALSE RFE values;
false_values
```

### For the detailed statistics, using Statsmodel to build a Model;

#### Building Model - A

```python
train_X_A = train_X[true_values]                      # Creating a dataframe with RFE selected variables;
```

```python
import statsmodels.api as sm

train_X_A = sm.add_constant(train_X_A)                # Now adding a Constant Variable here;
```

```python
LR = sm.OLS(train_Y,train_X_A).fit()                  # Running the Linear Model;

LR.summary()                                          # Summary of our Linear Model;
```

Our Analysis:

- R-squared value = 0.844
- F-statistics = 191
- Adj. R-square = 0.840
- Variables p-values
 - p-value of Nov: 0.144 and Jan: 0.054 are high

### Variable's VIF check

```python
train_X_New = train_X_A.drop(["const"], axis = 1)                   # Dropping the Constant Column;
```

- Calculating the VIFs for our New Model

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor 


# Fetching the values for VIFs in this new model;


VIF = pd.DataFrame()
X = train_X_New
VIF['Features'] = X.columns
VIF['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
VIF['VIF'] = round(VIF['VIF'], 2)
VIF = VIF.sort_values(by = "VIF", ascending = False)
VIF
```

```python
plt.figure(figsize = (18, 12))        # Checking the Correlation Coefficients to observe highly correlated Variables;
sns.heatmap(train_X_New.corr(), annot = True, cmap = "Blues")
plt.show()
```

## Train Data's Residual Analysis

- Now with the help of a Histogram, we will check how the error terms are distributed.

```python
train_Y_prediction = LR.predict(train_X_LR)


fig = plt.figure()

plt.scatter (train_Y, train_Y_prediction)               # Using Scatter plot with train_Y and train_Y_prediction;

fig.suptitle ("train_Y VS train_Y_prediction", fontsize = 18)    

plt.xlabel("train_Y", fontsize = 14)      

plt.ylabel("train_Y_prediction", fontsize = 14)   

plt.show()
```

```python
A = train_Y - train_Y_prediction

fig = plt.figure()

sns.distplot (A, bins = 20)                                          # Histogram for Error terms;

fig.suptitle ('Error Terms', fontsize = 20)                          # Heading;

plt.xlabel ('Errors', fontsize = 18)    

plt.show()
```
